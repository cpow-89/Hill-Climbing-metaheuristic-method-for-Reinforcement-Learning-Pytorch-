# Hill Climbing method for finding the weights in a Reinforcement Learning Problem 

### Hill Climbing
- iterative local search algorithm for optimization
- we use it here to optimize the weights of a policy network to solve a Reinforcement Learning problem

### Algorithm:
a.) set $initial\_weights$ to random values<br>
b.) set $best\_weights$ to $initial\_weights$<br>
c.) evaluate $best\_weights$ on one episode of the given task and collect the $episode\_reward$<br>
d.) set $best\_weight\_episode\_reward$  to $episode\_reward$<br>
e.) generate $new\_weights$ based on the following formular:<br>

$new\_weights = best\_weights + \sigma * gaussian\_noise$<br>
$\sigma$ = noise weight factor<br>
$gaussian\_noise$ = random sample from a normal(gaussian) distribution with shape of our weights<br>

f.) evaluate $new\_weights$ on one episode of the given task and collect the $episode\_reward$<br>
g.) if the $episode\_reward$ of $new\_weights$ is higher than the $best\_weight\_episode\_reward$ <br>

set $best\_weights$ to $new\_weights$<br>
set $best\_weight\_episode\_reward$  to $episode\_reward$<br>
else do nothing<br>

h.) repeat e-g until the task is eventually solved

# Examples

### MountainCarContinuous v0

- MountainCarContinuous-v0 defines "solving" as getting average reward of 90.0 over 100 consecutive trials.

```python
import json
import os
import gym
import numpy as np
from agent import HillClimbing
import session
import helper
```

```python
with open(os.path.join(".", "configs", "MountainCarContinuous_v0.json"), "r") as read_file:
    config = json.load(read_file)

np.random.seed(101)
agent = HillClimbing(config)  
    
if config["is_training"]:
    total_reward = session.train(agent, config)
    agent.save()
    helper.plot_scores(total_reward)
else:
    env = gym.make(config["env_name"])
    agent.load()
    session.evaluate(agent, env, config)
    env.env.close()
    env.close()
```
